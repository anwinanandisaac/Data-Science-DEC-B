**# QUESTION 5:**The management needs your help to give bonus points to the top 25% of
students based on their math score, so how will you help the management
to achieve this. **bold text**

```python
# Calculate the score threshold for the top 25% of students based on math scores
math_threshold = data_df['math score'].quantile(0.75)

# Identify the top performers based on math scores
top_performers = data_df[data_df['math score'] >= math_threshold]

# Assign bonus points to the top performers
bonus_points = 10   # You can adjust the number of bonus points as needed
top_performers['bonus_points'] = bonus_points

# Display the top performers and their bonus points
print("Top Performers and Bonus Points:")
print(top_performers[['math score', 'bonus_points']])
```

another method using .iloc

```python
# Sort the DataFrame based on math scores in descending order
data_df_sorted = data_df.sort_values(by='math score', ascending=False)

# Calculate the index position for the top 25% of students
top_25_percent_index = int(len(data_df_sorted) * 0.25)

# Select the top performers using iloc
top_performers = data_df_sorted.iloc[:top_25_percent_index]

# Assign bonus points to the top performers
bonus_points = 10  # You can adjust the number of bonus points as needed
top_performers['bonus_points'] = bonus_points

# Display the top performers and their bonus points
print("Top Performers and Bonus Points:")
print(top_performers[['math score', 'bonus_points']])
```
