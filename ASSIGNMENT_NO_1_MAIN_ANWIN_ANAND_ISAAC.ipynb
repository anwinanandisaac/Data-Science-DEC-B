# QUESTION 5.: PERFORM MEANINGFUL VISUALIZATIONS USING THE DATASET.

```python
# Scatter plot using matplotlib: GRAPH PLOTTED PW(PETAL WIDTH) ALONG Y-AXIS vs PL(PETAL LENGTH) ALONG X-AXIS
#                           ONE OUT OF THREE:
plt.scatter(data.PL,data.PW)         #SYNTAX FOR SCATTERPLOTTING
plt.xlabel('PL(petal length)')       #NAMING THE X-AXIS
plt.ylabel('PW(petal width)')        #NAMING THE Y-AXIS
plt.title('GRAPH BETWEEN PL vs PW')  #NAMING THE TITLE
```

```python
# Histogram using seaborn: THE BELOW GRAPH REPRESENTS HOW MANY OF THE PETALS OF THE GIVEN DATA ARE SAME/IDENTICAL:
#              TWO OUT OF THREE

sns.histplot(x=data.PL,color='black')
```

```python
# Scatter plot using matplotlib: GRAPH PLOTTED PW(PETAL WIDTH) ALONG Y-AXIS vs PL(PETAL LENGTH) ALONG X-AXIS
#                           ONE OUT OF THREE:
plt.scatter(data.PL,data.SL)         #SYNTAX FOR SCATTERPLOTTING
plt.xlabel('PL(petal length)')       #NAMING THE X-AXIS
plt.ylabel('SL(sepel length)')        #NAMING THE Y-AXIS
plt.title('GRAPH BETWEEN PW vs SL')  #NAMING THE TITLE
```
